```python
root_dir = "librispeech_test_clean/LibriSpeech/test-clean"

rows = []

for speaker_id in os.listdir(root_dir):
    speaker_path = os.path.join(root_dir, speaker_id)

    if not os.path.isdir(speaker_path):
        continue

    for chapter_id in os.listdir(speaker_path):
        chapter_path = os.path.join(speaker_path, chapter_id)

        if not os.path.isdir(chapter_path):
            continue

        trans_files = [f for f in os.listdir(chapter_path) if f.endswith(".trans.txt")]

        for trans_file in trans_files:
            trans_path = os.path.join(chapter_path, trans_file)

            with open(trans_path, "r") as f:
                lines = f.readlines()

            for line in lines:
                parts = line.strip().split(" ", 1)
                audio_id = parts[0]
                transcript = parts[1]

                audio_path = os.path.join(chapter_path, audio_id + ".flac")

                if os.path.exists(audio_path):
                    rows.append({
                        "audio_path": audio_path,
                        "transcript": transcript
                    })

df = pd.DataFrame(rows)

df_400 = df.head(400)
df_400.to_csv("librispeech_300.csv", index=False)

print(df_400.head())
print("Total samples saved:", len(df_400))
```

```python
def normalize_text(text):
  text = text.upper()
  text = re.sub(r"[^A-Z0-9\s]", "", text)
  text = re.sub(r"\s+", " ", text).strip()
  return text
```

```python
from transformers import Wav2Vec2Processor, Wav2Vec2ForCTC

model_name = "facebook/wav2vec2-base-960h"

processor = Wav2Vec2Processor.from_pretrained(model_name)
w2v_model = Wav2Vec2ForCTC.from_pretrained(model_name)

w2v_model.to("cuda")

print("wav2vec2 loaded")
```

```python
from datasets import Dataset

train_dataset = Dataset.from_pandas(dev_df.reset_index(drop=True))
eval_dataset = Dataset.from_pandas(eval_df.reset_index(drop=True))

print(train_dataset)
print(eval_dataset)
```

```python
import librosa
```

```python
def load_audio(batch):
    audio, sr = librosa.load(batch["audio_path"], sr=16000)
    batch["speech"] = audio
    batch["sampling_rate"] = sr
    batch["target_text"] = batch["transcript"].upper()
    return batch
```

```python

train_dataset = train_dataset.map(load_audio)
eval_dataset = eval_dataset.map(load_audio)

print(train_dataset[0].keys())
```

```python
def prepare_batch(batch):
    # audio input
    audio_inputs = processor(
        audio=batch["speech"],
        sampling_rate=16000
    )

    # text label
    label_inputs = processor(
        text=batch["target_text"]
    )

    batch["input_values"] = audio_inputs["input_values"][0]
    batch["labels"] = label_inputs["input_ids"]

    return batch
```

```python
train_dataset = train_dataset.map(
    prepare_batch,
    remove_columns=train_dataset.column_names
)

eval_dataset = eval_dataset.map(
    prepare_batch,
    remove_columns=eval_dataset.column_names
)

print(train_dataset[0].keys())
```

```python
from dataclasses import dataclass
from typing import Dict, List, Union


@dataclass
class DataCollatorCTCWithPadding:
    processor: object
    padding: Union[bool, str] = True

    def __call__(self, features):
        input_features = [
            {"input_values": feature["input_values"]}
            for feature in features
        ]

        label_features = [
            {"input_ids": feature["labels"]}
            for feature in features
        ]

        batch = self.processor.pad(
            input_features,
            padding=self.padding,
            return_tensors="pt"
        )

        labels_batch = self.processor.pad(
            labels=label_features,
            padding=self.padding,
            return_tensors="pt"
        )

        labels = labels_batch["input_ids"].masked_fill(
            labels_batch.attention_mask.ne(1), -100
        )

        batch["labels"] = labels

        return batch

data_collator = DataCollatorCTCWithPadding(processor=processor)
```

```python
import evaluate

wer_metric = evaluate.load("wer")

def compute_metrics(pred):
    pred_logits = pred.predictions
    pred_ids = np.argmax(pred_logits, axis=-1)

    pred.label_ids[pred.label_ids == -100] = processor.tokenizer.pad_token_id

    pred_str = processor.batch_decode(pred_ids)
    label_str = processor.batch_decode(pred.label_ids, group_tokens=False)

    wer_value = wer_metric.compute(
        predictions=pred_str,
        references=label_str
    )

    return {"wer": wer_value}
```

```python
w2v_model.freeze_feature_encoder()
```

```python
from transformers import TrainingArguments

training_args = TrainingArguments(
    output_dir="./wav2vec2-finetuned-librispeech-400",
    per_device_train_batch_size=4,
    per_device_eval_batch_size=4,
    gradient_accumulation_steps=2,
    eval_strategy="epoch",
    save_strategy="epoch",
    num_train_epochs=10,
    learning_rate=1e-5,
    warmup_steps=50,
    fp16=True,
    logging_steps=10,
    save_total_limit=2,
    report_to="none"
)
```

```python
from transformers import Trainer

trainer = Trainer(
    model=w2v_model,
    args=training_args,
    train_dataset=train_dataset,
    eval_dataset=eval_dataset,
    processing_class=processor,
    data_collator=data_collator,
    compute_metrics=compute_metrics
)

print("Trainer ready")
```

```python
trainer.train()
```

```python
trainer.evaluate()
```

```python
eval_df.to_csv("whisper_eval_results.csv", index=False)
trainer.save_model("./wav2vec2_finetuned")
processor.save_pretrained("./wav2vec2_finetuned")
```

```python
# 1) Evaluate once and store actual Wav2Vec2 WER
w2v_eval_metrics = trainer.evaluate()
wav2vec_avg_wer = w2v_eval_metrics["eval_wer"]
print("Fine-tuned Wav2Vec2 Average WER:", wav2vec_avg_wer)

# 2) Save model and processor
trainer.save_model("./wav2vec2_finetuned")
processor.save_pretrained("./wav2vec2_finetuned")
print("Saved model to ./wav2vec2_finetuned")

# 3) Make per-sample Wav2Vec2 predictions and WER column
```

```python
w2v_model.eval()

w2v_predictions = []
w2v_wers = []

def transcribe_wav2vec(audio_path):
    speech, sr = librosa.load(audio_path, sr=16000)
    inputs = processor(speech, sampling_rate=16000, return_tensors="pt", padding=True)
    input_values = inputs.input_values.to(w2v_model.device)

    with torch.no_grad():
        logits = w2v_model(input_values).logits

    pred_ids = torch.argmax(logits, dim=-1)
    pred_text = processor.batch_decode(pred_ids)[0]
    return pred_text

for _, row in eval_df.iterrows():
    true_text = row["transcript"].upper()
    pred_text = transcribe_wav2vec(row["audio_path"]).upper()

    sample_wer = wer(true_text, pred_text)
    w2v_predictions.append(pred_text)
    w2v_wers.append(sample_wer)

eval_df["wav2vec_predicted_text"] = w2v_predictions
eval_df["wav2vec_wer"] = w2v_wers

eval_df.to_csv("wav2vec_eval_results.csv", index=False)
print("Saved wav2vec_eval_results.csv")
print("Average per-sample Wav2Vec2 WER:", sum(w2v_wers) / len(w2v_wers))
display(eval_df[["audio_path", "transcript", "wav2vec_predicted_text", "wav2vec_wer"]].head())

wav2vec_avg_wer = sum(w2v_wers) / len(w2v_wers)

results = {
    "Model": ["Fine-tuned Wav2Vec2"],
    "Dataset": ["LibriSpeech eval split"],
    "WER": [wav2vec_avg_wer]
}

comparison_df = pd.DataFrame(results)
display(comparison_df)
```

```python
%%writefile app.py
import streamlit as st
import torch
import librosa
from transformers import Wav2Vec2ForCTC, Wav2Vec2Processor

MODEL_PATH = "wav2vec2_finetuned"

@st.cache_resource
def load_model():
    processor = Wav2Vec2Processor.from_pretrained(MODEL_PATH)
    model = Wav2Vec2ForCTC.from_pretrained(MODEL_PATH)
    model.eval()
    return processor, model

processor, model = load_model()

st.title("Speech to Text using Fine-tuned Wav2Vec2")

audio_file = st.file_uploader("Upload audio", type=["wav", "mp3", "flac"])

if audio_file is not None:
    st.audio(audio_file)

    speech, sr = librosa.load(audio_file, sr=16000)

    inputs = processor(
        speech,
        sampling_rate=16000,
        return_tensors="pt",
        padding=True
    )

    with torch.no_grad():
        logits = model(inputs.input_values).logits

    pred_ids = torch.argmax(logits, dim=-1)
    transcription = processor.batch_decode(pred_ids)[0]

    st.subheader("Transcription")
    st.write(transcription)
```

```python
import gradio as gr
import torch
import librosa
from transformers import Wav2Vec2ForCTC, Wav2Vec2Processor

MODEL_PATH = "wav2vec2_finetuned"

processor = Wav2Vec2Processor.from_pretrained(MODEL_PATH)
model = Wav2Vec2ForCTC.from_pretrained(MODEL_PATH)
model.eval()

def transcribe(audio_path):
    speech, sr = librosa.load(audio_path, sr=16000)
    inputs = processor(speech, sampling_rate=16000, return_tensors="pt", padding=True)

    with torch.no_grad():
        logits = model(inputs.input_values).logits

    pred_ids = torch.argmax(logits, dim=-1)
    return processor.batch_decode(pred_ids)[0]

demo = gr.Interface(
    fn=transcribe,
    inputs=gr.Audio(type="filepath"),
    outputs="text",
    title="Speech to Text using Fine-tuned Wav2Vec2"
)

demo.launch(share=True)
```
